```python
from pathlib import Path

target_folder = Path.cwd().parent / "Data" / "Bluebins"
new_base_name = "bluebins"
extensions = {'.jpg', '.jpeg', '.png', '.bmp', '.tiff'}

if not target_folder.exists():
    print(f"Error: The folder '{target_folder}' does not exist.")
else:
    # Get and sort files
    files = sorted([f for f in target_folder.iterdir() if f.is_file() and f.suffix.lower() in extensions])

    for count, file_path in enumerate(files, start=1):
        # Create new filename with 4-digit padding
        new_filename = f"{new_base_name}_{count:04d}{file_path.suffix}"
        new_file_path = target_folder / new_filename
        
        # Rename
        file_path.rename(new_file_path)
        print(f"Renamed: {file_path.name} -> {new_filename}")

    print("\nRenaming complete!")
```
