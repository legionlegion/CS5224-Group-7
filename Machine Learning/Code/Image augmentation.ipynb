```python
import albumentations as A
import cv2
import os
from pathlib import Path

# Define paths
input_path = Path(r"C:\Users\shenl\OneDrive\Documents\NUS\MS2\CS5224-Cloud Computing\Project\Machine Learning\Data\Bluebins\bluebins_001.jpg")
output_dir = Path(r"C:\Users\shenl\OneDrive\Documents\NUS\MS2\CS5224-Cloud Computing\Project\Machine Learning\Data\Image augmentation test")

# Ensure output directory exists
output_dir.mkdir(parents=True, exist_ok=True)

# Load image
image = cv2.imread(str(input_path))
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Define augmentations
transformations = {
    "rotation": A.Rotate(limit=45, p=1),
    "horizontal_flip": A.HorizontalFlip(p=1),
    "shift_scale_rotate": A.ShiftScaleRotate(shift_limit=0.1, scale_limit=0.1, rotate_limit=0, p=1),
    "brightness_contrast": A.RandomBrightnessContrast(p=1),
    "hsv_jitter": A.HueSaturationValue(hue_shift_limit=20, sat_shift_limit=30, val_shift_limit=20, p=1),
    "grayscale": A.ToGray(p=1),
    "gaussian_noise": A.GaussNoise(var_limit=(10.0, 50.0), p=1),
    "random_erasing": A.CoarseDropout(max_holes=8, max_height=32, max_width=32, p=1)
}

# Apply and save
base_name = input_path.stem
for name, transform in transformations.items():
    augmented = transform(image=image)["image"]
    save_path = output_dir / f"{base_name}_{name}.jpg"
    
    # Save as BGR for OpenCV compatibility
    cv2.imwrite(str(save_path), cv2.cvtColor(augmented, cv2.COLOR_RGB2BGR))
    print(f"Transformation '{name}' saved to {save_path}")
```
